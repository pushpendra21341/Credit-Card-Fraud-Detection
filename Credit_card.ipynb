Printing dataset using GUI

```python
import pandas as pd
from tkinter import *
from tkinter import ttk

df = pd.read_csv('creditcard.csv')

page_size = 100
total_pages = (len(df) + page_size - 1) // page_size

def update_treeview(page_num):
    start_index = (page_num - 1) * page_size
    end_index = min(start_index + page_size, len(df))
    page_data = df.iloc[start_index:end_index]
    
    for row in treeview.get_children():
        treeview.delete(row)
    
    for _, row in page_data.iterrows():
        treeview.insert('', 'end', values=row.tolist())

def navigate_pages(direction):
    global current_page
    if direction == 'next' and current_page < total_pages:
        current_page += 1
    elif direction == 'prev' and current_page > 1:
        current_page -= 1
    
    update_treeview(current_page)

root = Tk()
root.title("Excel Viewer")

columns = df.columns.tolist()
treeview = ttk.Treeview(root, columns=columns, show='headings')

for col in columns:
    treeview.heading(col, text=col)

current_page = 1
update_treeview(current_page)

prev_button = Button(root, text='Previous', command=lambda: navigate_pages('prev'))
prev_button.pack(side=LEFT)
next_button = Button(root, text='Next', command=lambda: navigate_pages('next'))
next_button.pack(side=LEFT)

treeview.pack()
root.mainloop()
```
